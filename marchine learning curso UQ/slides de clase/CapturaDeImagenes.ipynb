```python
# Inicializar un arreglo para almacenar los nombres de los archivos .png
imagenes = []
etiquetas = []
# Recorrer todos los archivos en la carpeta
for archivo in os.listdir(ruta+'/papel'):
    # Verificar si el archivo es de tipo .png
    if archivo.endswith('.png'):
        # Agregar el nombre del archivo al arreglo
        img = load_img(ruta+'/papel/'+archivo)
        img = img_to_array(img, dtype=np.uint8)
        imagenes.append(img)
        etiquetas.append('papel')

# Recorrer todos los archivos en la carpeta
for archivo in os.listdir(ruta+'/piedra'):
    # Verificar si el archivo es de tipo .png
    if archivo.endswith('.png'):
        img = load_img(ruta+'/piedra/'+archivo)
        img = img_to_array(img, dtype=np.uint8)
        imagenes.append(img)
        etiquetas.append('piedra')

# Recorrer todos los archivos en la carpeta
for archivo in os.listdir(ruta+'/tijeras'):
    # Verificar si el archivo es de tipo .png
    if archivo.endswith('.png'):
        img = load_img(ruta+'/tijeras/'+archivo)
        img = img_to_array(img, dtype=np.uint8)
        imagenes.append(img)
        etiquetas.append('tijeras')
```

```python
imagenes = np.array(imagenes)
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(etiquetas)
imagenes.shape, labels.shape
```
